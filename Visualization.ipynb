### Importing necessary libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("data_for_visualization.csv")
```

```python
data
```

```python
%matplotlib inline
plt.style.use("ggplot")
plt.figure(num=None, figsize=(8,6),dpi=120)
plt.title("Microsoft Stocks")
plt.scatter(data.year,data.high,label="Max stock price achieved");
plt.legend();
plt.xlabel("Year")
plt.ylabel("Stock Price (USD)")
plt.scatter(data.year,data.low,label="Min stock price encountered");
plt.legend();
plt.show()
plt.close()
```

#### Observation:
* Highest variance between max price and min price was observed in year 2020 indicating high market activity.
* Lowest variance between max price and min price was observed in year 2005 indicating low market activity.

```python
plt.figure(num=None, figsize=(8,6),dpi=120)
plt.title("Microsoft Stocks")
plt.plot(data["year"],data["average price"],linewidth=2.0,linestyle='-',color='b',alpha=0.5,marker='o',label="Average Stock Price");
plt.legend();
plt.xlabel("Year")
plt.ylabel("Stock Price (USD)")
```

#### Observation:
* Major decline in stock price is observed between year 2000 and 2005
* Steep rise in stock price is observed between years 2015 and 2020

```python
plt.figure(num=None, figsize=(8,6),dpi=120)
plt.title("Microsoft Stocks")
#didn't wanted to print for 1999
plt.bar(data["year"][1:],data["growth"][1:],label="Growth per Year");
plt.legend();
plt.xlabel("Year")
plt.ylabel("Stock Price (USD)")
```

#### Observation:
* Major decline in growth is observed in year 2000 indicating loss
* High growth rate is observed in year 2001 and 2020 indicating profit
